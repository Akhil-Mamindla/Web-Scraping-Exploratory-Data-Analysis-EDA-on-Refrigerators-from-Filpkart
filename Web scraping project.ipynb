```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36',
         'content-type': 'text/plain'}
brand=[]
capacity=[]
ratings=[]
price=[]
doors=[]
stars=[]
for i in range(1,51):          ## in 1 page 24 refrigerators  total 46 page , so range taken 1,46
    url=f"https://www.flipkart.com/search?q=refrigerator&sid=j9e%2Cabm%2Chzg&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=refrigerator%7CRefrigerators&requestId=24ff476e-0b6e-4e88-9170-4d00e348f9e0&as-searchtext=refri&page={i}"
    response=requests.get(url,headers=header)
    if response.status_code==200:
        soup=BeautifulSoup(response.text)
        
        for ele in soup.find_all("div",class_="jIjQ8S"):
         ## Brand
            m=ele.find("div",class_="RG5Slk").text
            m1=re.findall("^\w+",m)[0] 
            brand.append(m1)
        
         ## price
            y=ele.find("div",class_="hZ3P6w DeU9vF").text
            price.append(y)

         ## ratings
            r=ele.find("div",class_="MKiFS6")
            if r:
               ratings.append(r.text)
            else:
                ratings.append(np.nan)


         ## doors
            d=re.findall("\w+\sDoor",m)
            if d:
                doors.append(d[0])
            else:
                doors.append(np.nan)


         ## stars
            s=re.findall("(\d\s)Star",m)
            if s:
                stars.append(s[0])
            else:
                stars.append(np.nan)
                
         ##  capacity
            c=re.findall("(\d+\s)L",m)
            if c:
                capacity.append(c[0])
            else:
                capacity.append(np.nan)


    else:
         print("something went wrong")
            
                 
```

```python
## Removing NullValues from dataset
```

```python
## removing null values of starts
df["Stars"]=df["Stars"].fillna(df["Stars"].mode().values[0])
```

```python
## removing null values of Capacity
df["Capacity"]=df["Capacity"].fillna(df["Capacity"].mode().values[0])
```

```python
df.Capacity.isnull().sum()
```

```python
## removing null values of Doors
df["Doors"]=df["Doors"].fillna(df["Doors"].mode().values[0])
```

```python
df.Doors.isnull().sum()
```

```python
## removing null values of Ratings
df["Ratings"]=df["Ratings"].fillna(df["Ratings"].mode().values[0])
```

```python
df.Ratings.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["Price"]=df["Price"].apply(lambda x:re.sub("[^\d+]","",x))
```

```python
df["Price"]=df["Price"].astype("Int64")
```

```python
df["Capacity"] = pd.to_numeric(df["Capacity"], errors="coerce").astype("Int64")
```

```python
df["Stars"]= pd.to_numeric(df["Stars"], errors="coerce").astype("Int64")
```

```python
df["Ratings"]= pd.to_numeric(df["Ratings"], errors="coerce").astype("float")
```
